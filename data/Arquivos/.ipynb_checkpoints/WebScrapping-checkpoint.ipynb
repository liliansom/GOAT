```python
with open("CRPAG.html", "r", encoding="utf-8") as arquivo:
          site = BeautifulSoup(arquivo, "html.parser")

titulo = site.body
print(site.prettify())
```

```python
site.find_all("div", attrs={"class": "large-6 columns"})
```

```python
nome = site.find_all("h2", attrs={"class": "content-box-headline"})

for elemento in nome:
    texto_completo = elemento.text.strip()
    texto_dividido = texto_completo.split(' ', 1)
    primeiro_caractere = int(texto_dividido[0].replace('x', ''))
    texto = texto_dividido[1]
    for i in range(primeiro_caractere):
        if i == primeiro_caractere-1:
            print(texto, end="\n")
        else:
            print(texto, end="\n ")
```

```python
# construindo este rolê!!

titulo_ano = site.find_all("td", attrs={"class":"erfolg_table_saison zentriert"})  
titulo = site.find_all("td", attrs={"class":"no-border-links"})
celulas = []

if titulo_ano and titulo:
    for id in range(len(titulo_ano)):
        elemento3 = titulo_ano[id].text
        elemento1 = titulo[id].text
        if not elemento1:
            print(f'{elemento3.strip()}')
        else:
            print(f'{elemento3.strip()}, {elemento1.strip()}')
```
